#**1.Quantum Information**

Quantum information is a theory of computation based on the principles of quantum mechanics. Unlike traditional information processing theory, quantum information utilizes features of quantum mechanics such as the uncertainty principle, superposition, and quantum entanglement. Quantum information represents information using quantum bits or qubits. These qubits can exist in superposition states where they can be both 0 and 1 simultaneously. Quantum information has applications in areas such as quantum computing, quantum communication, and quantum cryptography. Research in this field aims to go beyond the limitations of traditional computers and communication systems.

**1.1. Using Quantum Information States**
-----------------------------------

Quantum information states are fundamental concepts used to describe the state of a system in quantum mechanics. These states define the properties and behavior of a quantum system. Quantum information states are typically defined on quantum bits (qubits) or other fundamental units of quantum systems.

In quantum mechanics, a system can be in a certain state at a particular time, and this state can be in superposition until a measurement is made on the system. These states are often expressed mathematically as vectors. For example, a qubit is represented by a state vector, which specifies the state in which the qubit exists.

The most common quantum information states include:

1. **Statevector**: A vector that specifies the complete state of a quantum system, encompassing the states of all qubits in the system.

2. **Density Matrix**: A matrix used to describe mixed states of a quantum system. The density matrix provides a complete statistical description of the quantum system.

3. **Stabilizer State**: A representation method used to simulate certain types of quantum error-correcting codes. These states are expressed using stabilizer circuits.

Quantum information states play a significant role in areas such as quantum computing, quantum communication, and quantum cryptography, forming the foundation for the design and utilization of quantum computers.

| Class Name        | Description                                |
|-------------------|--------------------------------------------|
| Statevector       | Represents a state vector                  |
| DensityMatrix     | Represents a density matrix                |
| StabilizerState   | Simulation of stabilizer circuits          |

**1.1.1. Statevector**

**Description**

- **Briefly, the Statevector class:** is used to represent the state of a quantum system. This class maintains all possible states of quantum bits (qubits) in a system in vector form.

- **Let's elaborate on this definition a bit more.** The Statevector class is an **n-dimensional complex vector** representing the complete state of a quantum system at a particular time. Each dimension corresponds to a qubit and is used to describe the state of a qubit. The state of a qubit is typically expressed with an amplitude vector. For instance, the state of a single qubit over the basis states |0⟩ and |1⟩ is a combination of the probability amplitudes: α|0⟩ + β|1⟩, where α and β are complex numbers.

**Applications**

- **In short, the Statevector class:** is widely used in quantum computing algorithms, circuit simulations, and quantum information processing applications. It is particularly prevalent in quantum circuit simulations and the analysis of quantum computing algorithms. Statevectors can be used to track the evolution of a quantum circuit and understand how a specific input state transforms into an output state.

**Features**

- This class contains a vector of complex numbers representing a series of probabilities. Each component represents the probability of a specific qubit state. The Statevector class is crucial for simulating computations in quantum circuits.

- In a broader sense, the Statevector typically encompasses all possible states of a qubit array. For instance, all possible states in a system consisting of n qubits are represented by a vector of size 2^n. The Statevector class provides a range of functions and operations including qubit addition, subtraction, combining states, and operations with measurement operators.



**1.1.2. DensityMatrix**

**Description**

- The DensityMatrix class is used to represent the density matrix of a quantum system. The density matrix encompasses the probabilities of all quantum states in a system and their relationships with each other.

*In a broader sense...*

- The DensityMatrix class is a matrix that defines the complete state of a quantum system. This matrix includes the states of a single qubit or a group of qubits and their relationships with each other. The density matrix is used to describe both pure and mixed quantum states.

**Applications**

- The DensityMatrix class is particularly used in the analysis of complex quantum systems and experiments related to quantum mechanics. It provides detailed information about the state of a specific qubit or group of qubits. Additionally, the density matrix is used to analyze correlations and entanglement states between qubits.

**Features**

- The density matrix is used to model non-pure states of a quantum system. This class is used to describe correlations between qubits, degrees of mixture, and entanglement states.

*In a broader sense...*

- The density matrix typically represents a mixture of states of a single qubit or a group of qubits. Various properties of this matrix are utilized to characterize the complexity and correlations of the quantum system. For example, the trace of the density matrix can be used to determine whether the system is pure or mixed.

**A. Usage of the Statevector Class**
------------------

As discussed above, in its simplest terms, the Statevector class represents a quantum state vector and includes functionality for initialization and operations on the state vector.

Now let's delve a bit deeper into this definition with examples.

```python
#1.
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
```

```python
#2.
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
```

```python
#3.
statevector = Statevector(qc)
print(statevector)
```

So what does this result tell us?

* This Statevector represents the state of a 2-qubit quantum system. **dims=(2,2)** Here, this expression denotes a system with 2 qubits and 2 possibilities.

-------------------------
For a 3-qubit system, dims=(2,2,2)

-------------------------


* In the given Statevector, there are **4 elements:**
 1. 0.70710678+0.j,
 2. 0+0.j,
 3. 0+0.j,
 4. 0.70710678+0.j.

 Each element represents a qubit state. In a state vector representing a quantum system, there are 2 possible states for each qubit. These states are denoted as |0⟩ and |1⟩. Therefore, in a 2-qubit system, there are a total of 4 possible states.

* As understood from the above results, the probability amplitudes of the qubits are equal, indicating that they are in a Bell State. This means there is maximum entanglement between the two qubits. This implies that changes in the state of one qubit are dependent on the other qubit, and measurements made on one qubit instantly determine the state of the other qubit.

Another way we can create a Statevector, different from above, is by passing a normalized complex vector.

**Let's discuss an example right away...**

```python
#1.
import numpy as np
from qiskit.quantum_info import Statevector
```

```python
#2.
statevector = Statevector([1, 0, 0, 1] / np.sqrt(2))
print(statevector.data)
```

There's a small thing we need to explain here.

* **[1,0,0,1] / np.sqrt(2)**

Here, a normalized complex vector is created. This vector contains the values [1, 0, 0, 1] and is divided by np.sqrt(2). This operation is done to normalize the vector's norm to 1. This vector represents the state of a qubit pair. The first two elements represent the states for the first qubit, while the last two elements represent the states for the second qubit.

So, what does **Normalizing** mean?

"Normalizing" means adjusting the size of a vector to 1. Mathematically, to normalize a vector, you need to divide each element of the vector by the norm (length) of the vector.

* The norm of a vector is typically referred to as the Euclidean norm, and it is the square root of the sum of the squares of the elements of the vector. The Euclidean norm measures the length or magnitude of a vector.

* In quantum mechanics, the norm of a state vector should be equal to 1. This is necessary to accurately represent the probabilities of quantum systems. Thus, normalizing a state vector means dividing each element of the vector by its norm, so that the length of the vector becomes 1. This ensures proper representation of state vectors used in quantum mechanics.

**B. Using the DensityMatrix Class**
-------------------------

As discussed earlier, the DensityMatrix class represents a quantum density matrix.

Here, we will now examine examples related to this class.

```python
#1.
from qiskit import QuantumCircuit
from qiskit.quantum_info import DensityMatrix
```

```python
#2.
qc = QuantumCircuit(2)

#3.
qc.h(0)
qc.cx(0, 1)
qc.z(1)
```

```python
#4.
dens_mat = DensityMatrix(qc)
print(dens_mat.data)
```

Now here, we need to discuss the following:

* While DensityMatrix contains a complex matrix,
* Statevector contains a complex vector.

This situation above allows:
* DensityMatrix to represent an ensemble of two or more quantum states,
* Statevector to represent only pure states.

**Now let's examine these with an example.**

```python
#1
from qiskit.quantum_info import DensityMatrix,Operator

#2.
dens_mat = 0.5*DensityMatrix.from_label('11') + 0.5*DensityMatrix.from_label('+0')
print(dens_mat.data)
```

```python
#3.
tt_op = Operator.from_label('TT')
dens_mat = dens_mat.evolve(tt_op)
dens_mat.draw('city')
```

```python
#4.
print(dens_mat.probabilities())
```

Now let's measure 1000 times and use `sample_counts` to sample the probability distribution.

```python
#5.
print(dens_mat.sample_counts(1000))
```
