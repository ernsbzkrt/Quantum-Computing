**3. Quantum Measurements and Visualization of States**
-------------------------------------------

Visualizing quantum states and measurements is a method that helps make the abstract concepts of quantum physics more accessible. This is often done using vectors or probability distributions. Visualization emphasizes the uncertainty and probabilistic nature of quantum systems. In this chapter, we will focus on these topics.

**3.1. Visualizing Measurement Outcomes**

In quantum computing processes, when a specific quantum circuit or algorithm is run, the results are often obtained in the form of measurement outcomes.

Quantum computing processes are built on uncertainty and probability. Therefore, to understand the results of a quantum computing process, it is important to visualize the probabilities of different measurement outcomes. Visualizing measurement outcomes shows the likelihood of measuring a particular quantum state and the distribution of these measurements.

These visualizations are crucial for understanding whether a quantum circuit is functioning correctly, diagnosing errors, and evaluating the performance of algorithms. Additionally, they can be used to comprehend the complexity of quantum computing processes. As a result, visualizing measurement outcomes is a critical tool for understanding and evaluating the results of quantum computing processes.

**Qiskit includes the** `plot_histogram()` **method to visualize measurement outcomes. This method takes a dictionary containing measurement counts and plots a bar graph for each basis state.**

*Let's create an example related to this scenario below.*

```python
#1.
pip install qiskit
```

```python
#2.
pip install qiskit-aer
```

```python
#3.
from qiskit import QuantumCircuit,transpile
from qiskit.visualization import plot_histogram
from qiskit_aer import Aer

#
qc = QuantumCircuit(2)

#
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

#
backend = Aer.get_backend("aer_simulator")
tqc = transpile(qc, backend)
job = backend.run(tqc, shots=1024)

#
result = job.result()
counts = result.get_counts(tqc)
plot_histogram(counts)
```

The table below shows a list of commonly used **plot_histogram parameters**. We typically customize with these parameters.

| Parameter Name | Description                                       |
|--------------|-----------------------------------------------|
| data         | A dictionary or list of dictionaries containing measurement counts.            |
| figsize      | A tuple containing the figure size in inches.                   |
| color        | A string or list of strings for bar colors.                   |
| legend       | A list of strings to label the data. The number of entries should match the number of dictionaries in the data parameter. |
| bar_labels   | A Boolean value that causes each bar to be labeled with probability values.            |
| title        | A string to title the graph.                           |

**3.2. Visualization of Quantum States**

The visualization of quantum states is typically done through vectors, matrices, or more complex structures. For example, the state of a quantum bit (qubit) can be visualized as a point on a Bloch sphere. This visualization helps us understand the superposition or the probability of being in a certain state of a qubit.

Visualization of quantum states is also used to understand specific quantum phenomena such as quantum entanglement (the complex relationship between interconnected qubits). For instance, the entanglement state between two qubits can be visualized as a relationship between a pair of qubits known as an EPR pair.

Visualizing quantum states can facilitate the learning process for beginners in quantum science by making quantum mechanical principles more tangible and understandable. Moreover, these visualizations play a significant role in the analysis and understanding of complex quantum systems, as well as in the development of quantum algorithms and protocols.

**Qiskit** includes several functions to visualize state vectors and density matrices. These functions are provided in the table below:

| Function                   | Description                                                               |
|--------------------------|------------------------------------------------------------------------|
| plot_state_qsphere()     | Visualizes the state vector in a three-dimensional spherical representation called Qsphere.    |
| plot_state_city()        | Represents the state vector with a 2D visualization resembling a cityscape.       |
| plot_bloch_multivector()| Visualizes state vectors for one or multiple qubits on the Bloch sphere. |
| plot_state_hinton()      | Displays the state vector as a Hinton diagram.                |
| plot_state_paulivec()    | Visualizes the state vector as a combination of Pauli operators.                 |

These functions can be used to visualize quantum states in different manners, aiding users in better understanding quantum systems.

**Now let's delve into each of them in detail.**

**3.2.1. The plot_state_qsphere() Function**

This function is commonly used in the following scenarios:

1. **Single Qubit States:** The state of a qubit can be visualized on the Qsphere. In this case, the state of the qubit is visualized globally, allowing for a better understanding of the relationship between the qubit's classical and quantum states.

2. **Bell States and Entanglement:** Entanglement states between two or more qubits can be visualized on the Qsphere. Since special cases like Bell states play a crucial role in quantum communication and algorithms, visualizing such states can aid in understanding these protocols.

3. **States of Quantum Circuits:** The state of a specific quantum circuit can be visualized on the Qsphere to understand the output of the circuit. This can help in tracking and analyzing quantum computation processes.

While it can be used for these three scenarios, this method is particularly excellent for **visualizing multi-qubit quantum states.**

Let's examine this function with an example.

```python
#1.
from qiskit.visualization import plot_state_qsphere
from math import pi
```

```python
#2.
backend = Aer.get_backend("aer_simulator_statevector")

#3.
qc = QuantumCircuit(3)
qc.rx(pi, 0)
qc.ry(pi/8, 2)
qc.swap(0, 2)
qc.h(0)
qc.cp(pi/2,0, 1)
qc.cp(pi/4, 0, 2)
qc.h(1)
qc.cp(pi/2, 1, 2)
qc.h(2)
qc.save_statevector()
```

```python
#4.
tqc = transpile(qc, backend)
job = backend.run(tqc)
result = job.result()
statevector = result.get_statevector()
plot_state_qsphere(statevector)
```

**Let's discuss the above figure in detail:**

It depicts a Q-sphere, where each basis state is represented by a point. The size of each point is proportional to the measurement probability, and the color of a point corresponds to its phase. We should consider that the basis states are placed on the poles and meridians of the sphere, starting with all zeros at the top and all ones at the bottom, based on the Hamming weight (the number of ones in the basis states).

In the table below, we can see the most commonly used parameters related to the **plot_state_qsphere** function. These parameters are used to customize the generated graph.

| Parameter Name    | Description                                                                                                                                                                              |
|-------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| state             | A NumPy type representing a pure or mixed quantum state, containing complex numbers, such as Statevector, DensityMatrix, or ndarray.                                                    |
| figsize           | A tuple containing the figure size in inches.                                                                                                                                            |
| show_state_labels | A Boolean value to show labels for each basis state.                                                                                                                                     |
| show_state_phases | A Boolean value to show phases for each basis state.                                                                                                                                     |
| use_degrees       | A Boolean value indicating whether to use degrees for phase values in the plot.                                                                                                          |

**3.2.2. The plot_state_city() Function**

This function is particularly preferred for single qubit states. It visualizes the state vector in a 2D representation. Each qubit state is represented in a row, and each column represents different states of the qubit.

The plot_state_city function is useful for simple and understandable visualizations of quantum circuits. For example, it can be used to visualize the effects of applying a Hadamard gate on a qubit or to understand a superposition state on a qubit. This visualization enables a more intuitive understanding of qubit states.

Let's examine this function with an example.

```python
#1.
from qiskit.quantum_info import DensityMatrix, Operator
from qiskit.visualization import plot_state_city
```

```python
#2.
dens_mat = 0.5*DensityMatrix.from_label('11') + \
0.5*DensityMatrix.from_label('+0')
tt_op = Operator.from_label('TT')
dens_mat = dens_mat.evolve(tt_op)

#3.
plot_state_city(dens_mat)
```

In the table below, we can see the most commonly used parameters related to the **plot_state_city** function. These parameters are used to customize the generated graph.

| Parameter Name | Description                                                                                         |
|----------------|-----------------------------------------------------------------------------------------------------|
| state          | Statevector, DensityMatrix, or ndarray (NumPy type) containing complex numbers representing a pure or mixed quantum state |
| title          | A string to title the graph                                                                          |
| figsize        | A tuple containing the figure size in inches                                                          |
| color          | A list containing two elements for colors of real and imaginary 3D bars                                |
| alpha          | A float containing the desired transparency for 3D bars                                                |

**3.2.3. The plot_bloch_multivector() Function**

The **plot_bloch_multivector() function** takes a state vector or density matrix and represents it on one or more Bloch spheres. This function is useful for visualizing the states of quantum circuits, especially those containing single or multiple qubits.

The **plot_bloch_multivector() function** is used to show the states of qubits on the Bloch sphere. It can be utilized to observe if a qubit is in a superposition state, in a Bell state, or how its state changes after applying a gate.

Let's examine this function with an example right away.

```python
#1.
from qiskit import QuantumCircuit,transpile
from qiskit.visualization import plot_bloch_multivector
from math import pi
```

```python
#2.
backend = Aer.get_backend("aer_simulator_statevector")
```

```python
#3.
qc = QuantumCircuit(3)
qc.rx(pi, 0)
qc.ry(pi/8, 2)
qc.swap(0, 2)
qc.h(0)
qc.cp(pi/2,0, 1)
qc.cp(pi/4, 0, 2)
qc.h(1)
qc.cp(pi/2, 1, 2)
qc.h(2)
qc.save_statevector()
```

```python
#4.
tqc = transpile(qc, backend)
job = backend.run(tqc)

#5.
result = job.result()
statevector = result.get_statevector()
plot_bloch_multivector(statevector)
```

The above figure depicts a **Bloch sphere** for each qubit in a quantum state. **However, the arrow on the Bloch sphere does not always reach the surface of the sphere.** This occurs typically in the following cases:

- When there is entanglement (complex interconnection) between qubits. In an entangled state, as the states of qubits are correlated, the state of a single qubit cannot be isolated, and thus the arrow on the Bloch sphere may not reach the surface of the sphere.

- When the state is a mixed state. A mixed quantum state is not a pure state but rather a mixture of different pure states. In this case, the arrow on the Bloch sphere may also not reach the surface of the sphere because the state being a mixture, the states of the qubits cannot be precisely placed at a particular point.

**3.2.4. The plot_state_hinton() Function**

The Hinton diagram is used to visually represent the complexity of a quantum state and the magnitudes of its components.

This function is typically used to visualize complex quantum states, particularly useful for cases involving multiple qubits or mixed states. The Hinton diagram displays both the real and imaginary parts of a state vector or density matrix, providing more information about the structure of the quantum state.

Let's examine it with an example right away.

```python
#1.
from qiskit.quantum_info import DensityMatrix,Operator
from qiskit.visualization import plot_state_hinton

#2.
dens_mat = 0.5*DensityMatrix.from_label('11') + 0.5*DensityMatrix.from_label('+0')
tt_op = Operator.from_label('TT')
dens_mat = dens_mat.evolve(tt_op)

#3.
plot_state_hinton(dens_mat)
```

**3.2.5. The plot_state_paulivec() Function**

The plot_state_paulivec() function is used to visualize the expectation values of a quantum state on the Pauli matrices as a sparse bar graph. This function can be used to visualize rotations or average spin values of the quantum state in the x, y, and z directions.

In particular, the plot_state_paulivec() function is useful when you want to understand and visualize the expectation values of various Pauli operators for a quantum state. This function is valuable for analyzing the states in quantum computers and evaluating the performance of quantum circuits.

Let's examine it with an example right away.

```python
#1.
from qiskit.quantum_info import DensityMatrix,Operator
from qiskit.visualization import plot_state_paulivec
```

```python
#2.
dens_mat = 0.5*DensityMatrix.from_label('11') + 0.5*DensityMatrix.from_label('+0')
tt_op = Operator.from_label('TT')
dens_mat = dens_mat.evolve(tt_op)

#3.
plot_state_paulivec(dens_mat)
```
